```python
# =========================
# 1. DATA PREPROCESSING
# =========================

import pandas as pd
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords

# Load datasets
resume_df = pd.read_csv("Resume.csv")
job_df = pd.read_csv("training_data.csv")

# Check columns
print(resume_df.columns)
print(job_df.columns)

# Clean column names
resume_df.columns = resume_df.columns.str.strip()
job_df.columns = job_df.columns.str.strip()

# Rename columns
resume_df.rename(columns={'Resume_str': 'Resume'}, inplace=True)
job_df.rename(columns={'job_description': 'Job_Description'}, inplace=True)

# Stopwords
stop_words = set(stopwords.words('english'))

# Cleaning function
def clean_text(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

# Apply cleaning
resume_df['clean_resume'] = resume_df['Resume'].apply(clean_text)
job_df['clean_job'] = job_df['Job_Description'].apply(clean_text)

# Save cleaned files
resume_df.to_csv("cleaned_resumes.csv", index=False)
job_df.to_csv("cleaned_jobs.csv", index=False)

print("Preprocessing Done ")
print(resume_df.head())
```

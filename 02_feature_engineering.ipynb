```python
# =========================
# 2. FEATURE ENGINEERING
# =========================

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

# Load cleaned data
resume_df = pd.read_csv("cleaned_resumes.csv")
job_df = pd.read_csv("cleaned_jobs.csv")

#  FIX: remove NaN
resume_df['clean_resume'] = resume_df['clean_resume'].fillna('')
job_df['clean_job'] = job_df['clean_job'].fillna('')

# Combine text
all_text = list(resume_df['clean_resume']) + list(job_df['clean_job'])

# TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)
vectorizer.fit(all_text)

resume_vectors = vectorizer.transform(resume_df['clean_resume'])
job_vectors = vectorizer.transform(job_df['clean_job'])

# Save vectors
pickle.dump(resume_vectors, open("resume_vectors.pkl", "wb"))
pickle.dump(job_vectors, open("job_vectors.pkl", "wb"))

print("Feature Engineering Done ")
```

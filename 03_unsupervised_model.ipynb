```python
# =========================
# 3. MATCHING MODEL + RANKING
# =========================

import pickle
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load vectors
resume_vectors = pickle.load(open("resume_vectors.pkl", "rb"))
job_vectors = pickle.load(open("job_vectors.pkl", "rb"))

# Compute similarity
similarity_matrix = cosine_similarity(resume_vectors, job_vectors)

matches = []
top_matches = []

for i in range(len(similarity_matrix)):
    scores = similarity_matrix[i]
    
    # Best match
    best_idx = scores.argmax()
    best_score = scores[best_idx]
    matches.append((i, best_idx, best_score))
    
    #  Top-3 Ranking (IMPORTANT)
    sorted_idx = scores.argsort()[::-1]
    top3 = sorted_idx[:3]
    
    for j in top3:
        top_matches.append((i, j, scores[j]))

# DataFrames
match_df = pd.DataFrame(matches, columns=["Resume_ID", "Job_ID", "Score"])
top3_df = pd.DataFrame(top_matches, columns=["Resume_ID", "Job_ID", "Score"])

# Save
match_df.to_csv("matches.csv", index=False)
top3_df.to_csv("top3_matches.csv", index=False)

print("Matching Done ")
print(match_df.head())
print(top3_df.head())
```

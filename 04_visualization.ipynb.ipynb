```python
# =========================
# 4. VISUALIZATION
# =========================

import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("matches.csv")

scores = df['Score']

# Histogram
plt.hist(scores, bins=50)
plt.title("Similarity Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()

# Line plot
plt.plot(scores[:50])
plt.title("Top 50 Resume Matching Scores")
plt.xlabel("Index")
plt.ylabel("Score")
plt.show()

# Top-3 distribution
top3_df = pd.read_csv("top3_matches.csv")

plt.hist(top3_df['Score'], bins=50)
plt.title("Top-3 Match Score Distribution")
plt.show()

print("Visualization Done ")
```
